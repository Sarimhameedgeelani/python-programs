# CREATING ARRAYS WITH DIFFERENT METHODS


```python
#np.array:creating araay from lists/touples  eg
arr=np.array([1,2,3])
print(arr)

#np.zeros: creating array with zeros  rg
arr=np.zeros((2,3))
print(arr)

#np.ones : crating array from ones  eg
arr=np.ones((3,3))
print(arr)

#np.empty: creates uninitialized array  eg
arr=np.empty((2,3))
print(arr)
```

# reshaping arrays:


```python
#change dimentions
import numpy as np
arr=np.array([[1,2],
              [3,4]])
reshaped= arr.reshape(1,4)
print(reshaped)
```

```python
#changing size(in place)
arr=np.array([[1,2],
              [3,4]])
resized= arr.resize(1,3)
print(resized)
```

```python
#transpose elements
arr=np.array([[1,2],
              [3,4]])
transposed=arr.T
print(transposed)
```

```python
#flatten aray(returns 1d array)
arr=np.array([[1,2],
              [3,4]])
flattened= arr.flatten()
print(flattened)
```

```python
#ravel returns 1d view (faster)
arr=np.array([[1,2],
              [3,4]])
view= arr.ravel()
print(view)
```

```python
#flat interator over elements
arr=np.array([[1,2],
              [3,4]])
flat=list(arr.flat)
print(flat)
```

# IMAGE AS ARRAY

-images in numpy are represented as 3D arrays
#image ,array , shape(height, weadth and channels)

-height: number pf pixels vertically.
-weadth: number of fixels horizontaly.
-channels: RGB(3 channes)or grauscale(1 channel).
-values: 0-255(for unit 8 images)


```python
#read and display image
import matplotlib.pyplot as plt
import numpy as np

img= plt.imread(r"C:\Users\DELL\OneDrive\Pictures\spiderman 2.jpg")
plt.imshow(img)
plt.axis("on")
plt.title("original image")
plt.show
```

```python
#crop image
crop=img[200:600,200:700]
plt.title("croped image")
plt.imshow(crop)
```

```python
#convert to grayscale
gray = np.mean(img, axis=2)

plt.imshow(gray, cmap="gray")
plt.axis("on")
plt.title("Grayscale Image")
plt.show()
```

```python
#downscale image
small = img[::2, ::2]   # skip every 2 pixels

plt.imshow(small)
plt.axis("off")
plt.title("Resized (Downscaled)")
plt.show()
```

```python
#increase brightness
bright = img + 5
bright = np.clip(bright, 0, 255)

plt.imshow(bright.astype(np.uint8))
plt.axis("off")
plt.title("Brightened Image")
plt.show()
```

```python
#increase darkness
dark = img - 50
dark = np.clip(dark, 0, 255)

plt.imshow(dark.astype(np.uint8))
plt.axis("off")
plt.title("Darkened Image")
plt.show()
```

```python
#extract color channels
red = img.copy()
red[:, :, 1] = 0
red[:, :, 2] = 0

plt.imshow(red)
plt.axis("off")
plt.title("Red Channel")
plt.show()
```

```python
#blur image
kernel = np.ones((5,5,3)) / 25
blur = img.copy()

blur[1:-1,1:-1] = (
    img[:-2,:-2] + img[:-2,1:-1] + img[:-2,2:] +
    img[1:-1,:-2] + img[1:-1,1:-1] + img[1:-1,2:] +
    img[2:,:-2] + img[2:,1:-1] + img[2:,2:]
) / 9

plt.imshow(blur.astype(np.uint8))
plt.axis("off")
plt.title("Blurred Image")
plt.show()
```
